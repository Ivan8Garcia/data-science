## **¿Cuales son los valores promedio de alquiler por tipo de propiedad?**

```python
datos.head()
```

```python
datos["Valor"].mean()
```

```python
##usaremos groupby para agrupar primero los valores por la columna TIPO
datos.groupby("Tipo").mean(numeric_only=True)
```

```python
## especificamos  entre corchetes la columna VALOR para sacar el promedio solamente a esa columna
datos.groupby("Tipo")["Valor"].mean()
```

```python
## ahora agregamos otro corchete para covertirlo a DATA FRAME
datos.groupby("Tipo")[["Valor"]].mean().sort_values("Valor")
```

```python
df_tipo_precio= datos.groupby("Tipo")[["Valor"]].mean().sort_values("Valor")
df_tipo_precio.plot(kind="barh", figsize=(12,7), color="red")
```

```python
datos.Tipo.unique()
```

```python
inmuebles_comerciales=['Conjunto Comercial/Sala', 'Edificio Completo',
       'Tienda/Salón', 'Casa Comercial',
       'Terreno Estándar', 'Cochera/Estacionamiento',
       'Galpón/Depósito/Almacén', 'Tienda en Centro Comercial', 'Hotel', 'Loteo/Condominio',
       'Industria']
```

```python
datos.query("@inmuebles_comerciales in Tipo")
```

```python
datos.query("@inmuebles_comerciales not in Tipo")
```

```python
df=datos.query("@inmuebles_comerciales not in Tipo")
df.head()
```

```python
df_tipo_precio= df.groupby("Tipo")[["Valor"]].mean().sort_values("Valor")
df_tipo_precio.plot(kind="barh", figsize=(12,7), color="red")
```
