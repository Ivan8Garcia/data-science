##**Leyendo una consulta SQL**

```python
# SELECT* significa seleccionar todo
query= "SELECT * FROM clientes WHERE Categoria_de_renta = 'Empleado' "
```

```python
#todo el codigo lo guardamos en una varible llamada empleados, para leer el sql invocamos a la variable query de arriba y
# en esta version se agrega text para leerla, despues la conectamos a nuestro motor, que es engine.connect()
empleados = pd.read_sql(sql=text(query), con=engine.connect())
```

```python
#comprobamos que es un DATAFRAME
type(empleados)
```

```python
#VISUALIZAMOS EL DATA FRAME de solo "empleados", invocandolo
empleados
```

**generaremos una nueva tabla, empezando siempre de nuestro data frame, que se llama EMPLEADOS**

```python
#aqui se creó una tabla nueva llamada empleados porque si usamos la palabra "clientes", estaremos borrando la base de datos original
empleados.to_sql("empleados", con=engine.connect(), index= False)
```

```python
pd.read_sql_table("empleados", con=engine.connect())
```

Es hora de que pongas a prueba los conocimientos desarrollados durante la clase. Disponemos de un archivo en formato CSV, el cual se obtuvo de la página Información de Salud (TABNET) - DATASUS. Este archivo contiene los gastos hospitalarios públicos de cada una de las Unidades de la Federación Brasileña desde enero de 2008 hasta marzo de 2021.

Su desafío aquí será leer este archivo usando la función read_csv de la biblioteca Pandas. Se deben agregar algunos parámetros para que la lectura se realice correctamente. Entonces, aquí hay algunos consejos:

1.Asegúrese de que el archivo CSV esté separado por una coma o punto y coma.

2.La codificación del archivo es ISO-8859-1.

3.Las primeras tres líneas del archivo se pueden ignorar, ya que el encabezado sólo comienza en la cuarta línea.

4.Las últimas 9 líneas también se pueden ignorar, ya que son sólo información sobre dónde se tomaron los datos.

5.Para eliminar las últimas líneas es necesario agregar el parámetro engine='python'.

Para leer el archivo podemos comenzar almacenando la dirección en una variable llamada url:

```python
url = 'datos_sus.csv'
```

Luego podemos leer la url con la función read_csv y agregar los parámetros:

encoding='ISO-8859-1': tipo de codificación de caracteres del archivo. En este caso, es la codificación ISO-8859-1 la que es común en los archivos en idioma portugués.

sep=';': separador de los valores del archivo. En este caso, es el punto y coma (;).

skiprows=3: indica que las primeras tres líneas del archivo deben ignorarse ya que no contienen datos relevantes.

skipfooter=9: Indica que las últimas nueve líneas del archivo deben ignorarse ya que no contienen datos relevantes.

engine='python': el motor utilizado para leer el archivo es Python.

Esto es necesario cuando se utiliza el parámetro skipfooter, ya que el motor predeterminado no admite esta opción.

```python
datos = pd.read_csv(url, encoding='ISO-8859-1', sep=';', skiprows=3, skipfooter = 9, engine='python')
```

**RESOLUCION DEL DESAFIO**

Para crear la base de datos local es necesario ejecutar los siguientes comandos. Este código consiste en importar la biblioteca **sqlalchemy** y crear un objeto engine que representa una conexión a una base de datos **SQLite** en memoria. El método **create_engine** se utiliza para crear el objeto engine y especificar la base de datos que se utilizará. En este caso, la base de datos es una base de datos SQLite en memoria, que se crea usando **sqlite:///:memory:**

```python
import sqlalchemy
from sqlalchemy import create_engine, MetaData, Table, inspect, text
engine = create_engine('sqlite:///:memory:')
```

A continuación, podemos leer los datos con la función read_csv y escribirlos en el formato de la base de datos con la función to_sql:

```python
import pandas as pd
archivo = 'clientes_banco.csv'
datos = pd.read_csv(archivo)
datos.to_sql('clientes', engine, index=False)
```

Después de crear la tabla, podemos obtener un DataFrame con la función read_sql_table:



```python
pd.read_sql_table('clientes', con=engine.connect())
```

Actualizar el registro del cliente ID 6840104 cuyo rendimiento anual cambió a 300000:

El siguiente código actualiza el rendimiento anual del ID de cliente 6840104 en la tabla **clientes**. La variable **query** contiene una declaración SQL que actualiza ese registro. Luego, la declaración **engine.connect()** se usa para crear una conexión a la base de datos SQLite. Finalmente, la declaración **.execute(text(query))** se utiliza para ejecutar la declaración SQL contenida en la variable **query** y actualizar el registro en la tabla **clientes**.

```python
from sqlalchemy.exc import SQLAlchemyError
query = 'UPDATE clientes SET Rendimiento_anual=300000.0 WHERE ID_Cliente=6840104'
try:
  r_set=engine.connect().execute(text(query))
except SQLAlchemyError as e:
  print(e)
else:
  print("#Registros actualizados: ",r_set.rowcount)
```

Después de ejecutar este código, puede verificar si la actualización realmente ocurrió leyendo la tabla "clientes"

```python
pd.read_sql_table('clientes', con=engine.connect())
```

Para utilizar esta cláusula, primero debe especificar el nombre de la tabla en la que se insertará el registro. A continuación, debemos pasar entre paréntesis los nombres de las columnas donde se agregarán los nuevos valores. **Es importante recordar que el orden de los nombres debe ser el mismo orden en el que se pasarán los valores.**

Después de especificar las columnas, usamos la cláusula **VALUES** para agregar los nuevos valores. Nuevamente debemos pasar entre paréntesis los valores a insertar en el mismo orden en que se especificaron las columnas.
